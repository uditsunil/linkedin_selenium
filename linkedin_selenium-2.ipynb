```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
import selenium .webdriver.support.expected_conditions as EC
from selenium.common.exceptions import TimeoutException
path = '/Applications/chromedriver'
driver = webdriver.Chrome(path)
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import csv



class ClickAndSendKeys():
    def test(self):
        baseUrl = "https://www.linkedin.com"
        x=['Data Scientist','Web Developer','Web Designer','Content Writer']
        driver.get(baseUrl)
        log = driver.find_element_by_xpath('//*[@class="login-email"]').send_keys('linkedin_email_id')
        pas = driver.find_element_by_xpath('//*[@class="login-password"]').send_keys('linkedin_password'')
        submit = driver.find_element_by_xpath('//*[@id="login-submit"]').click()
        
        def link():
            
            texts=[tmp.lower() for tmp in x]
            name = [s.replace(" ", "-") for s in texts]
            for i in range(len((name))):
                try:
                
                    driver.get('http://www.linkedin.com/jobs/search/?keywords={}'.format(name[i]))
                    driver.implicitly_wait(10)
                    jobtitle = driver.find_elements_by_xpath('//*[contains(@class,"job-card-search__title lt-line-clamp lt-line-clamp--multi-line ember-view")]')
                    company = driver.find_elements_by_xpath('//*[contains(@class,"job-card-search__company-name")]')
                    for i,j in zip(jobtitle,company):
                        newdf = pd.DataFrame({"jobtitle":[i.text],"company":[j.text]})
                        with open('linkedin.csv', 'a',newline = '') as f:
                            newdf.to_csv(f, line_terminator='\n',sep=',', index=False, header=False)
                except NoSuchElementException:
                    newdf = pd.DataFrame({"jobtitle":['none'],"company":['none']})
                    with open('linkedin.csv', 'a',newline = '') as f:
                        newdf.to_csv(f, line_terminator='\n',sep=',', index=False, header=False)
                        
                    driver.implicitly_wait(10)
                    driver.close()
                    
        link()


ff = ClickAndSendKeys()
ff.test()



```
